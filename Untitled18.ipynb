```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

data = {
    'Area': [1000, 1300, 1700, None, 2100, 1900, 1400, None, 2500, 1600],
    'Bedrooms': [2, 2, 3, 1, 4, 3, 2, 1, 5, 3],
    'Age': [8, 5, 4, 10, 2, None, 6, 12, 1, 5],
    'Rent': [12000, 16000, 23000, 10000, 32000, 27000, 18000, 9000, 38000, 21000]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)

print("\nDataset After Handling Missing Values:\n")
print(df)

X = df[['Area', 'Bedrooms', 'Age']]
y = df['Rent']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=10
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\nActual Rent:")
print(list(y_test))

print("\nPredicted Rent:")
print(y_pred)

mae = mean_absolute_error(y_test, y_pred)
print("\nMean Absolute Error:", round(mae, 2))

new_house = [[1800, 3, 4]]

predicted_rent = model.predict(new_house)

print("\nPredicted Rent for New House:", round(predicted_rent[0], 2))
```
